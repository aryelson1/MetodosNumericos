```python
def bissecao (a,b,f, tol = 1e-10, erro = 1):
    ant = a 
    ai = a
    bi = b
    i = 0
    iteracao = 0
    if ((np.sign(f(a)) * np.sign(f(b))) != -1 ):
        print("Não existe raiz no intervalo")
        return None
    else:
        while(erro > tol):
            medio = ((ai + bi)/2)
            if (f(medio) == 0):
                break
            elif (np.sign(f(ai)) * np.sign(f(medio)) == -1):
                bi = medio
            else:
                ai = medio
            erro = np.abs((medio - ant)/medio)
            i = i + 1
            ant = medio 
            iteracao +=1
    #return medio,iteracao // Caso da questão 5
    return medio

func = lambda x: 2 - np.exp(x)
print(bissecao(0,1,func))
```

```python
def interpolacao (a,b,f, tol = 1e-10, erro = 1):
    ant = a 
    ai = a
    bi = b
    i = 0
    iteracao = 0
    if ((np.sign(f(ai)) * np.sign(f(bi))) != -1 ):
        print("Não existe raiz no intervalo")
        return None
    else:
        while(erro > tol):
            sup = np.sign(f(bi)) * (ai - bi)
            inf = np.sign(f(ai)) - np.sign(f(bi))
            xa = bi - (sup/inf)
            if (f(xa) == 0):
                break
            elif (np.sign(f(a)) * np.sign(f(xa)) == -1):
                bi = xa
            else:
                ai = xa
            erro = np.abs((xa - ant)/xa)
            i = i + 1
            ant = xa
            iteracao += 1
    return [xa,iteracao]

func = lambda x: (-12*x**5) - (6*x**3) + 10 
print(interpolacao(0,1,func))
print(bissecao(0,1,func))
```
